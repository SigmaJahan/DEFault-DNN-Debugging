```python
from glob import glob
import os
import sys
import pandas as pd
import numpy as np
```

```python
#for buggy and correct files for CSV files
buggy_dir = "D:\\ICSE_Dataset\\CSVAllBuggy"
correct_dir = "D:\\ICSE_Dataset\\CSVAllCorrect"
balanced_dir = "D:\\ICSE_Dataset\\balancedCSVCorrectBuggy"

buggy_files = glob(os.path.join(buggy_dir, "*.csv"))
correct_files = glob(os.path.join(correct_dir, "*.csv"))
buggy_data = []
correct_data = []

for file in buggy_files:
    df = pd.read_csv(file)
    df['label'] = 'Buggy'
    buggy_data.append(df)

for file in correct_files:
    df = pd.read_csv(file)
    df['label'] = 'Correct'
    correct_data.append(df)

buggy_data = pd.concat(buggy_data, ignore_index=True)
correct_data = pd.concat(correct_data, ignore_index=True)
min_samples = min(len(buggy_data), len(correct_data))

buggy_sampled = buggy_data.sample(n=min_samples, random_state=42)
correct_sampled = correct_data.sample(n=min_samples, random_state=42)

balanced_data = pd.concat([buggy_sampled, correct_sampled], ignore_index=True)
if not os.path.exists(balanced_dir):
    os.makedirs(balanced_dir)

for i, row in balanced_data.iterrows():
    label = row['label']
    filename = f"{i}_{row['data_id']}_{label}.csv"
    row.to_frame().T.drop(columns=['label']).to_csv(os.path.join(balanced_dir, filename), index=False)

balanced_files = glob(os.path.join(balanced_dir, "*.csv"))

print(f"Balanced dataset created with {len(balanced_files)} files")
```

```python
#oversampling the correct files for CSV files (Original correct files number: 4797, will oversample to 7617)

import os
import shutil
import random

def oversample_files(original_dir, new_dir, target_count):
    csv_files = [f for f in os.listdir(original_dir) if f.endswith('.csv')]
    original_count = len(csv_files)
    if original_count >= target_count:
        print(f"Original file count ({original_count}) is already greater than or equal to target count ({target_count}).")
        return
    additional_files_needed = target_count - original_count
    sampled_files = random.choices(csv_files, k=additional_files_needed)
    os.makedirs(new_dir, exist_ok=True)
    for file in csv_files:
        shutil.copy(os.path.join(original_dir, file), os.path.join(new_dir, file))

    for idx, file in enumerate(sampled_files):
        new_filename = f"{idx + original_count + 1}_{file}"
        shutil.copy(os.path.join(original_dir, file), os.path.join(new_dir, new_filename))
    
    print(f"Oversampling complete. {target_count} files saved to {new_dir}.")

original_directory = "D:\\ICSE_Dataset\\CSVAllCorrect"
new_directory = "D:\\ICSE_Dataset\\OverSampledCorrect"
target_file_count = 7617
oversample_files(original_directory, new_directory, target_file_count)
```

```python
#Undersampled the correct files for CSV files (Original correct files number: 7617, will undersample to 4791)

import os
import shutil
import random

def undersample_files(original_dir, new_dir, target_count):
    csv_files = [f for f in os.listdir(original_dir) if f.endswith('.csv')]
    original_count = len(csv_files)
    if original_count <= target_count:
        print(f"Original file count ({original_count}) is already less than or equal to target count ({target_count}).")
        return
    sampled_files = random.sample(csv_files, target_count)
    os.makedirs(new_dir, exist_ok=True)
    for file in sampled_files:
        shutil.copy(os.path.join(original_dir, file), os.path.join(new_dir, file))
    
    print(f"Undersampling complete. {target_count} files saved to {new_dir}.")

original_directory = "D:\\ICSE_Dataset\\CSVAllBuggy"
new_directory = "D:\\ICSE_Dataset\\UnderSampledBuggy"
target_file_count = 4791
undersample_files(original_directory, new_directory, target_file_count)
```

```python
# changing the buggy file names
import os
import shutil

def rename_and_copy_files(original_dir, new_dir):
    os.makedirs(new_dir, exist_ok=True)

    for filename in os.listdir(original_dir):
        if filename.endswith('.csv'):
            parts = filename.split('_')
            if len(parts) == 4 and parts[3].endswith('.csv'):
                new_filename = f"{parts[0]}_{parts[1]}_{parts[3]}"
                old_filepath = os.path.join(original_dir, filename)
                new_filepath = os.path.join(new_dir, new_filename)
                shutil.copy(old_filepath, new_filepath)
                print(f"Copied and renamed: {filename} -> {new_filename}")
original_directory = "D:\\ICSE_Dataset\\UnderSampledBuggy"  
new_directory = "D:\\ICSE_Dataset\\CSVAllCorrectLevel1" 
rename_and_copy_files(original_directory, new_directory)
```

```python
import os

def add_index_to_files(directory):
    files = os.listdir(directory)
    for index, filename in enumerate(files, start=1):
        old_filepath = os.path.join(directory, filename)
        new_filename = f"{index}_{filename}"
        new_filepath = os.path.join(directory, new_filename)
        os.rename(old_filepath, new_filepath)
        print(f"Renamed: {filename} -> {new_filename}")

directory_path = "D:\\ICSE_Dataset\\balancedCorrectBuggy_Undersampled" 
add_index_to_files(directory_path)
```

```python
import os

def remove_second_number_from_filenames(directory):
    for filename in os.listdir(directory):
        parts = filename.split('_')
        if len(parts) > 3 and parts[1].isdigit() and parts[2].isdigit():
            new_filename = f"{parts[0]}_{parts[2]}_{parts[3]}"
            old_filepath = os.path.join(directory, filename)
            new_filepath = os.path.join(directory, new_filename)
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: {filename} -> {new_filename}")

directory_path = "D:\\ICSE_Dataset\\balancedCorrectBuggy_Oversampled" 
remove_second_number_from_filenames(directory_path)
```
