```python
#indexing DC++ and removing trailing number from file names in a directory
import os
import re

def index_files_in_directory(directory):
    files = sorted(os.listdir(directory))
    index = 0

    for filename in files:
        old_filepath = os.path.join(directory, filename)
        new_filename = f"{index}_{filename}"
        new_filepath = os.path.join(directory, new_filename)
        os.rename(old_filepath, new_filepath)
        print(f"Renamed: {filename} to {new_filename}")

        index += 1

def remove_trailing_number(directory):
    pattern = re.compile(r'_(\d+)\.(\w+)$')
    for filename in os.listdir(directory):
        match = pattern.search(filename)
        if match:
            new_filename = filename[:match.start()] + '.' + match.group(2)
            old_filepath = os.path.join(directory, filename)
            new_filepath = os.path.join(directory, new_filename)
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: {filename} to {new_filename}")


directory = "D:\\ICSE_Dataset\\test"
index_files_in_directory(directory)
remove_trailing_number(directory)
print("Indexing and trailing number removal done!!")
```

```python
#formatting DeepCrime filename and moving the smaller csv files to a separate directory
import os
import csv
import shutil

def copy_and_rename_files(source_directory, target_directory, file_extension, index):
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
    
    for filename in os.listdir(source_directory):
        if filename.endswith(file_extension):
            new_filename = filename.replace('deepmultifixrnn_', '').replace('deepmultifixFNN_', '').replace('correct_', '')
            index_mutated = new_filename.find('mutated')
            if index_mutated != -1:
                new_filename = new_filename[:index_mutated].rstrip('_') + file_extension
            else:
                new_filename = new_filename.replace(file_extension, '') + file_extension
            new_filename_with_index = f"{index}_{new_filename}"
            old_filepath = os.path.join(source_directory, filename)
            new_filepath = os.path.join(target_directory, new_filename_with_index)
            if not os.path.exists(new_filepath):
                shutil.copy2(old_filepath, new_filepath)
                print(f"Copied and renamed: {filename} to {new_filename_with_index}")
                index += 1

    print(f"Copying and renaming completed for {file_extension} files.")
    return index

def move_files_containing_word(directory, word, target_directory):
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)
    for filename in os.listdir(directory):
        if word in filename:
            old_filepath = os.path.join(directory, filename)
            new_filepath = os.path.join(target_directory, filename)
            if not os.path.exists(new_filepath):
                shutil.move(old_filepath, new_filepath)
                print(f"Moved file: {filename} to {target_directory}")

def move_small_csv_files(directory, target_directory, min_rows=50):
    if not os.path.exists(target_directory):
        os.makedirs(target_directory)

    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            filepath = os.path.join(directory, filename)
            try:
                with open(filepath, 'r') as csvfile:
                    reader = csv.reader(csvfile)
                    rows = list(reader)
            except csv.Error as e:
                print(f"Error reading {filename}: {e}")
                continue

            if len(rows) < min_rows:
                try:
                    new_filepath = os.path.join(target_directory, filename)
                    shutil.move(filepath, new_filepath)
                    print(f"Moved small CSV file: {filename} to {target_directory}")
                except PermissionError as e:
                    print(f"PermissionError moving {filename}: {e}")

source_directory_h5 = "D:\\ICSE_Dataset\\DC_H5"
source_directory_csv = "D:\\ICSE_Dataset\\DC_CSV"
target_directory_h5 = "D:\\ICSE_Dataset\\H5CorrectDeepCrime"
target_directory_csv = "D:\\ICSE_Dataset\\CSVCorrectDeepCrime"
target_directory_renamed_h5 = "D:\\ICSE_Dataset\\H5BuggyDeepCrime"
target_directory_renamed_csv = "D:\\ICSE_Dataset\\CSVBuggyDeepCrime"
small_csv_directory = "D:\\ICSE_Dataset\\smallCSVDeepCrime"

move_files_containing_word(source_directory_h5, 'original', target_directory_h5)
move_files_containing_word(source_directory_csv, 'original', target_directory_csv)

index = 0
index = copy_and_rename_files(source_directory_h5, target_directory_renamed_h5, '.h5', index)

index = 0
move_small_csv_files(source_directory_csv, small_csv_directory, min_rows=50)
index = copy_and_rename_files(source_directory_csv, target_directory_renamed_csv, '.csv', index)
print("Operations completed yay for now!!!")
```

```python
#renaming the correct files in a directory
import os
import re

def modify_filenames(directory):
    for filename in os.listdir(directory):
        match = re.match(r'(\d+)_deepmultifixFNN_(\d+)_correct.*\.(\w+)', filename)
        if match:
            index = match.group(1)
            number = match.group(2)
            extension = match.group(3)
            new_filename = f"{index}_{number}_correct.{extension}"
            old_filepath = os.path.join(directory, filename)
            new_filepath = os.path.join(directory, new_filename)
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: {filename} to {new_filename}")

directory = "D:\\ICSE_Dataset\\CSVCorrectDeepCrime"
modify_filenames(directory)
print("Filename modification completed.")
```

```python
#change the label name in Deep Crime
import os

rename_mapping = {
    'change_batch_size': 'HBS',
    'change_epochs': 'HNE',
    'change_learning_rate': 'HLR',
    'disable_batching': 'HDB',
    'change_activation_function': 'ACH',
    'change_optimisation_function': 'OCH',
    'change_loss_function': 'FLC',
    'change_dropout_rate': 'RCD',
    'remove_activation_function': 'RAF',
    'remove_bias': 'RBI',
    'change_weight_initialisation': 'CWI',
    'add_activation_function': 'AAF',
}

def rename_files(directory, extension):
    for filename in os.listdir(directory):
        if filename.endswith(extension):
            new_filename = filename
            for original, new in rename_mapping.items():
                if original in new_filename:
                    print(f"Found '{original}' in {filename}, replacing with '{new}'")
                    new_filename = new_filename.replace(original, new)
        
            if new_filename != filename:
                old_filepath = os.path.join(directory, filename)
                new_filepath = os.path.join(directory, new_filename)
                os.rename(old_filepath, new_filepath)
                print(f"Renamed: {filename} to {new_filename}")

directory_h5 = "D:\\ICSE_Dataset\\H5BuggyDeepCrime"
directory_csv = "D:\\ICSE_Dataset\\CSVBuggyDeepCrime"

rename_files(directory_h5, '.h5')
rename_files(directory_csv, '.csv')
print("Renaming completed.")
```

```python
#grouping the files in both H5 and CSV direcoties based on the label name

import os

category_mapping = {
    'HBS': 'Hyperparameter',
    'HNE': 'Hyperparameter',
    'HLR': 'Hyperparameter',
    'HDB': 'Hyperparameter',
    'ACH': 'Activation',
    'RAF': 'Activation',
    'AAF': 'Activation',
    'OCH': 'Optimization',
    'FLC': 'Loss',
    'RCD': 'Regularization',
    'LCO': 'Layer',
    'LCT': 'Layer',
    'LRM': 'Layer',
    'LAD': 'Layer',
    'LCN': 'Layer',
    'LKS': 'Layer',
    'LCF': 'Layer',
    'LCS': 'Layer',
    'LCD': 'Layer',
    'LCP': 'Layer',
    'RBI': 'Weight',
    'CWI': 'Weight',
}

def rename_files_by_category(directory):
    for filename in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, filename)):
            name, ext = os.path.splitext(filename)
            for key, value in category_mapping.items():
                if key in name:
                    new_name = name.replace(key, value)
                    new_filename = f"{new_name}{ext}"
                    os.rename(os.path.join(directory, filename), os.path.join(directory, new_filename))
                    print(f"Renamed: {filename} -> {new_filename}")
                    break 

directory_path_h5 = "D:\\ICSE_Dataset\\H5AllBuggy"
directory_path_CSV = "D:\\ICSE_Dataset\\CSVAllBuggy"

rename_files_by_category(directory_path_h5)
rename_files_by_category(directory_path_CSV)
```

```python
#counting the label and file distribution for both correct and buggy files
import os
import re
from collections import defaultdict, Counter


csv_category_counts = defaultdict(int)
h5_category_counts = defaultdict(int)
unique_number_counts = Counter()

def count_files_by_label(directories):
    for directory in directories:
        for filename in os.listdir(directory):
            if filename.endswith('.h5'):
                match = re.search(r'_(\D+)\.', filename)
                if match:
                    label = match.group(1)
                    h5_category_counts[label] += 1
            elif filename.endswith('.csv'):
                match = re.search(r'_(\D+)\.', filename)
                if match:
                    label = match.group(1)
                    csv_category_counts[label] += 1
                
                number_match = re.search(r'_(\d+)_', filename)
                if number_match:
                    number = number_match.group(1)
                    unique_number_counts[number] += 1


directories = [
    "D:\\ICSE_Dataset\\H5AllBuggy",
    "D:\\ICSE_Dataset\\CSVAllBuggy",
    "D:\\ICSE_Dataset\\H5AllCorrect",
    "D:\\ICSE_Dataset\\CSVAllCorrect",

]

count_files_by_label(directories)
print("CSV File Counts:")
for category, count in csv_category_counts.items():
    print(f"{category}: {count} files")

print("\nH5 File Counts:")
for category, count in h5_category_counts.items():
    print(f"{category}: {count} files")

print("\nUnique Number Distribution:")
for number, count in unique_number_counts.items():
    print(f"{number}: {count} files")
    
unique_number_counts = dict(unique_number_counts)
sorted_unique_number_counts = dict(sorted(unique_number_counts.items(), key=lambda item: item[1], reverse=True))
print("\nSorted Unique Number Distribution:")
for number, count in sorted_unique_number_counts.items():
    print(f"{number}: {count} files")

print("Counting completed.")
```
